```python
# pretrain, unbalance

data_accu = {
    "ResNet":{
        "accuracy origin": 0.79,
        "accuracy total": 0.76,
        "accuracy class 6": 0.49
    },
    "VGG":{
        "accuracy origin": 0.86,
        "accuracy total": 0.84,
        "accuracy class 6": 0.72
    },
    "Inception":{
        "accuracy origin": 0.75,
        "accuracy total": 0.71,
        "accuracy class 6": 0.38
    },
}

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
# draw accuracy
plt.figure(figsize=(8, 6))

# draw grouped bar plot with labels
barWidth = 0.25
r1 = np.arange(len(data_accu.keys()))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# bars, with numbers on the bar

plt.bar(r1, [data_accu[k]["accuracy origin"] for k in data_accu.keys()], width=barWidth, label="accuracy origin")
plt.bar(r2, [data_accu[k]["accuracy total"] for k in data_accu.keys()], width=barWidth, label="accuracy total")
plt.bar(r3, [data_accu[k]["accuracy class 6"] for k in data_accu.keys()], width=barWidth, label="accuracy class 6")


plt.xticks([r + barWidth for r in range(len(data_accu.keys()))], data_accu.keys())
#  add number

plt.legend()
plt.ylim(0, 1)
plt.ylabel("accuracy")
plt.xlabel("model")
plt.title("Accuracy of different models")

plt.savefig("accuracy.png", dpi=300)
plt.show()
```

```python
data_accu = {
    "balanced":{
        "accuracy pre trained": 0.76,
        "accuracy no pretrain": 0.63
    },
    "unbalanced":{
        "accuracy pre trained": 0.71,
        "accuracy no pretrain": 0.61
    },
}

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
# draw accuracy
plt.figure(figsize=(8, 6))

# draw grouped bar plot with labels
barWidth = 0.3
r1 = np.arange(len(data_accu.keys()))
r2 = [x + barWidth for x in r1]

# bars, with numbers on the bar
plt.bar(r1, [data_accu[k]["accuracy pre trained"] for k in data_accu.keys()], width=barWidth, label="accuracy pre trained")
plt.bar(r2, [data_accu[k]["accuracy no pretrain"] for k in data_accu.keys()], width=barWidth, label="accuracy no pretrain")

plt.xticks([r + barWidth/2  for r in range(len(data_accu.keys()))], data_accu.keys())

#  add number


plt.legend()
plt.ylim(0, 1)
plt.ylabel("accuracy")
plt.xlabel("balanced and unbalance - ResNet")
plt.title("Accuracy of ResNet")

plt.savefig("accuracy.png", dpi=300)
plt.show()
```
