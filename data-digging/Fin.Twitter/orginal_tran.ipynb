# 原始训练，无任何处理

```python
import pandas as pd
import jieba
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from lazypredict.Supervised import LazyClassifier

# 读取数据集
data = pd.read_csv('/home/train.csv')

# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['target'], test_size=0.2, random_state=42)

# 分词
X_train = X_train.apply(lambda x: ' '.join(jieba.cut(x)))
X_test = X_test.apply(lambda x: ' '.join(jieba.cut(x)))


# 向量化
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
X_train.shape
```

```python

# 训练模型
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(X_train.toarray(), X_test.toarray(),y_train,y_test)
models
# 评估模型性能
# y_pred = clf.predict(X_test)
# accuracy = accuracy_score(y_test, y_pred)
# print('Accuracy:', accuracy)
```

```python
import pickle

# 保存模型
with open('/home/model_raw_ori.pkl', 'wb') as f:
    pickle.dump(clf, f)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# save model to csv file

plt.figure(figsize=(10, 10))
sns.barplot(x='Accuracy', y='Model', data=models)
plt.title('Accuracy of Models')
plt.show()
```
